```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

# Filling Null Values

```python
np.random.seed(25)
df = DataFrame(np.random.rand(36).reshape(6,6))
df.loc[3:5,0] = np.nan
df.loc[1:4,5] = np.nan
df
```

```python
filled = df.fillna(method='ffill')
filled
```

# Counting Null Values in each Column

```python
df.isnull().sum()
```

# Filtering out Missing Values

```python
df.dropna() #removing rows containing null values
```

```python
df.dropna(axis=1) #removing columns containing null values
```

# Dropping Data

```python
df.drop([0,2]) #drops rows in the index 0 and 2
```

```python
df.drop([0,2], axis=1) #drops columns in the index 0 and 2
```

# sort Values

```python
df.sort_values(by=(4), ascending=[False]) #sort bu the column index 4 in descending order
```
